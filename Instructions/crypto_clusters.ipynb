```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
crypto_file_df = pd.read_csv('Resources/crypto_data.csv')
crypto_file_df.head()
```

```python
crypto_file_df.shape
```

```python
crypto_file_df.value_counts()
```

```python
crypto_file_df = crypto_file_df[crypto_file_df['IsTrading'] == True]
crypto_file_df.head()
```

```python
crypto_file_df = crypto_file_df.drop(columns ='IsTrading')
```

```python
crypto_file_df.isnull().sum()
```

```python
crypto_file_df = crypto_file_df.dropna()
crypto_file_df
```

```python
crypto_file_df.isnull().sum()
```

```python
crypto_file_df = crypto_file_df[crypto_file_df["TotalCoinsMined"]> 0]
crypto_file_df
```

```python
CoinName=pd.DataFrame(crypto_file_df['CoinName']).reset_index(drop=True)
CoinName
```

```python
crypto_file_df.drop(columns='CoinName', axis=1, inplace= True)
crypto_file_df
```

```python
crypto_file_df = crypto_file_df.drop(columns=['Unnamed: 0'])
crypto_file_df
```

```python
X = pd.get_dummies(crypto_file_df, columns=['Algorithm', 'ProofType'])
X
```

```python
# View column names 
X.columns
```

```python
scaler = StandardScaler()
crypto_scaled = scaler.fit_transform(X)
```

```python
len(X.columns)
```

```python
pca = PCA(n_components=4)

crypto_pca1 = pca.fit_transform(crypto_scaled)
```

```python
df_crypto_pca1 = pd.DataFrame(data=crypto_pca1, 
                             columns=["PC 1", "PC 2", "PC 3", "PC4"])
df_crypto_pca1.head()
```

```python
pca.explained_variance_ratio_
```

```python
pca2 = PCA(n_components=.90)

crypto_pca = pca2.fit_transform(crypto_scaled)
crypto_pca
```

```python
transformed_crypto_pca = pd.DataFrame(data=crypto_pca)
transformed_crypto_pca.head()
```

```python
pca.explained_variance_ratio_
```

```python
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='green')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)
```

```python
len(transformed_crypto_pca.columns)
```

```python
tsne = TSNE(learning_rate=35)
```

```python
tsne_features = tsne.fit_transform(transformed_crypto_pca)
tsne_features.shape
```

```python
x = tsne_features[:,0]

y = tsne_features[:,1]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
inertia = []
k = list(range(1, 11))

for i in k:
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(transformed_crypto_pca)
    inertia.append(km.inertia_)

elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
